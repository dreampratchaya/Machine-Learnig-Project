# Polynomial Regression

## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

**Note** เนื่องจากมีข้อมูลน้อยเลยไม่ต้องแบ่ง train กับ test

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

degree = 2
$$\textrm{y} = b_0 + b_1 \textrm{X} +b_2 X^2$$

degree = 3
$$\textrm{y} = b_0 + b_1 \textrm{X} +b_2 X^2 + b_3 X^3$$

degree = n
$$y = b_0 + b_1 X + b_2 X^2 + b_3 X^3 +...+ b_n X^n$$

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
a = {'real':y, 'pre':lin_reg_2.predict(X_poly)}
pd.DataFrame(a)
```

## Visualising the Linear Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth of Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth of Bluff (Polynomial Linear Regression), degree = 2')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

$ Note: le6 = 10^6$

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth of Bluff (Polynomial Linear Regression), degree = 4')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

**[[]] (array 2D) = [] ข้างนอกหมายถึง row, [] ข้างในหมายถึง colum**

**[[a,b],[c,d]]**

$\begin{bmatrix} a & b \\ c & d \end{bmatrix}$

```python
print(lin_reg.predict([[6.5]]))
```

## Predicting a new result with Polynomial Regression

```python
print(lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))
```
