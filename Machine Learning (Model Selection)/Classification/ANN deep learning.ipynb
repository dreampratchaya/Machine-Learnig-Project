```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
y = tf.keras.utils.to_categorical(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
#เนื่องจากพอ onw hot encode แล้วได้ 5 ตัวแปร output units จึงเท่ากับ 5
ann.add(tf.keras.layers.Dense(units = 5, activation = 'softmax'))
ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 200)
```

```python
y_pred = ann.predict(X_test)
print(np.round(y_pred,3))
```

```python
y_test
```

```python
ann.evaluate(X_test,y_test)
```

```python
tf.math.confusion_matrix(np.argmax(y_test, axis=1),np.argmax(y_pred, axis=1))
```

```python
# np.argmax จะดูที่ตำแหน่งที่ทำให้ค่ามากสุดอย่างเช่น 0,0,1,0,0 argmax ก็จะเท่กับ 2
np.argmax(y_test, axis=1)
```
