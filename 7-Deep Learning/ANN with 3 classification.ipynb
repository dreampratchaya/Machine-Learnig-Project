### Importing the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.datasets import load_iris
```

### Importing the dataset

```python
dataset = load_iris()
```

```python
dataset
```

```python
X = dataset.data
```

```python
y = dataset.target
```

```python
target_names = dataset.target_names
```

```python
y
```

### One Hot Encoding

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
y_test
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Building the ANN

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 3, activation = 'softmax'))
ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size = 1, epochs = 200)
```

```python
y_pred = ann.predict(X_test)
#np.round ปัดทศนิยม
print(np.round(y_pred,3))
```

```python
y_pred
```

```python
ann.evaluate(X_test,y_test)
```

### Making the Confusion Matrix

```python
tf.math.confusion_matrix(np.argmax(y_test, axis=1),np.argmax(y_pred, axis=1))
```

```python
data = {'Predict': target_names[np.argmax(y_pred, axis=1)], 'Actural': target_names[np.argmax(y_test, axis=1)]}
pd.DataFrame(data)
```
